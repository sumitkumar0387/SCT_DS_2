```python
sns.countplot(x="Survived",data=df,palette='colorblind')
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

#### 

```python
categorical_column=["Sex","Pclass","Embarked"]
for col in categorical_column:
    sns.countplot(x=col,data=df,palette="Set2")
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

```python
numerical_column=["Age","Fare"]
for col in numerical_column:
    sns.histplot(df[col],kde=True,color="Brown")
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
sns.countplot(x='Sex',hue="Survived",data=df,palette='Set2')
plt.title("Survival by Sex")
plt.show()
```

```python
sns.countplot(x="Pclass",hue="Survived",data=df)
plt.title("Survival count by Passenger Class")
plt.show()
```

```python
sns.countplot(x="Embarked",hue="Survived",data=df)
plt.title("Survival count by Embarkation Port")
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare',hue="Survived",data=df)
plt.title("Fare Distribution by Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

```python
sns.catplot(x='Pclass', hue='Survived', col='Sex', data=df, kind='count')
plt.suptitle("Survival Count by Sex and Passenger Class", y=1.25)
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df)
plt.title("Fare Distribution by Class and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

# ✅ Final Insights from Titanic Dataset

---

## 1. Survival Rate

- A significant portion of passengers did **not survive**.  
- **Females** had a much higher survival rate than males, showing that **gender played a crucial role** in survival.

---

## 2. Passenger Class Impact

- **First-class** passengers had the **highest survival rates**.  
- **Third-class** passengers had the **lowest**, indicating **socio-economic status influenced outcomes**.

---

## 3. Embarkation Port

- Most passengers boarded from **Southampton (S)**.  
- **Cherbourg (C)** passengers showed a **relatively higher survival proportion**.

---

## 4. Age Distribution

- **Younger passengers**, especially **children**, had better chances of survival.  
- The **20–40** age group was the **most common** on board.

---

## 5. Fare and Survival

- Passengers who paid **higher fares** had better survival rates.  
- Again, this reflects the **advantage of wealth/class**.

---

## 6. Gender and Class Combined

- **Female passengers in first and second class** had the **highest survival rates**.  
- **Male passengers in third class** were the **least likely to survive**.

---

## 7. Missing Data Handling

- The **‘Cabin’** column was **dropped** due to too many missing values.  
- **Median imputation** was used for missing **‘Age’** values.  
- **‘Embarked’** column’s missing values were **filled with the mode**.

---

## 8. Visualization Summary

- **Count plots** and **histograms** helped identify patterns in **categorical and numerical** variables.  
- **Box plots** revealed **fare distribution** differences between survivors and non-survivors.  
- **Scatter plots** and **cat plots** provided deeper insights into **multi-variable relationships**.

---

### 📌 These insights help in understanding the key factors that influenced survival during the Titanic disaster, laying the foundation for predictive modeling.
